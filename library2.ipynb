```python
import folium
from folium.plugins import HeatMap, HeatMapWithTime
from pymongo import MongoClient
import pandas as pd
from datetime import datetime as dt
```

```python
mongo = MongoClient(port=27017)
db = mongo.nasa_data
meteorites = db.meteorites.find()
```

```python
df = pd.DataFrame(meteorites)
df.head()
```

```python
drop_df = df.drop(['_id','nametype','fall','geolocation',':@computed_region_cbhk_fwbd',':@computed_region_nnqa_25f4'],axis = 1)
drop_df.head()
```

```python
df_count = drop_df.count()
df_count
```

```python
df_dropna = drop_df.dropna()
df_dropna.count()
```

```python
df_dropna.dtypes
```

```python

df_dropna['id'] = df_dropna['id'].astype(int)
df_dropna['mass'] = df_dropna['mass'].astype(float)
df_dropna['reclat'] = df_dropna['reclat'].astype(float)
df_dropna['reclong'] = df_dropna['reclong'].astype(float)
```

```python
df_dropna.dtypes
```

```python
df_dropna['year'] = pd.to_datetime(df_dropna['year'],errors='coerce')
df_dropna['year'] = df_dropna['year'].dt.strftime('%Y-%m-%d')
```

```python
yeardropdf = df_dropna.dropna()
yeardropdf.count()
```

```python
yeardropdf.rename(columns = {'name':'Name','recclass':'Class','mass':'Mass','year':'Year','reclat':'Latitude','reclong':'Longitude'},inplace=True)
yeardropdf
```

```python
yeardropdf.dtypes
```

```python
yeardropdf['Year'] = yeardropdf['Year'].sort_values(ascending=True)

data = []
for year, group in yeardropdf.groupby('Year'):
    data.extend([[row['Latitude'], row['Longitude'],row['Mass']] for _, row in group.iterrows()])
data
```
